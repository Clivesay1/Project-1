```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
median_path = "../Resources/Merged_sheet.csv"

median_df = pd.read_csv(median_path)

median_df.columns = median_df.columns.str.strip()

median_df.head()
```

```python


center_based_2018 = median_df['Median center-based price: 2018']
home_based_2018 = median_df['Median home-based price: 2018']
price_2018 = median_df['Median price: 2018']
center_based_2023 = median_df['Median center-based price: 2023']
home_based_2023 = median_df['Median home-based price: 2023']
price_2023 = median_df['Median price: 2023']
household_income_2018 = median_df['Median Household Income 2018']
household_income_2023 = median_df['Median Household Income 2023']

fig, axs = plt.subplots(4, 2, figsize=(14, 20))

# Histogram for center-based prices in 2018
axs[0, 0].hist(center_based_2018, bins=10, edgecolor='black')
axs[0, 0].set_title('Center-based Price Distribution in 2018')
axs[0, 0].set_xlabel('Price')
axs[0, 0].set_ylabel('Frequency')

# Histogram for home-based prices in 2018
axs[0, 1].hist(home_based_2018, bins=10, edgecolor='black')
axs[0, 1].set_title('Home-based Price Distribution in 2018')
axs[0, 1].set_xlabel('Price')
axs[0, 1].set_ylabel('Frequency')

# Histogram for median prices in 2018
axs[1, 0].hist(price_2018, bins=10, edgecolor='black')
axs[1, 0].set_title('Median Price Distribution in 2018')
axs[1, 0].set_xlabel('Price')
axs[1, 0].set_ylabel('Frequency')

# Histogram for center-based prices in 2023
axs[1, 1].hist(center_based_2023, bins=10, edgecolor='black')
axs[1, 1].set_title('Center-based Price Distribution in 2023')
axs[1, 1].set_xlabel('Price')
axs[1, 1].set_ylabel('Frequency')

# Histogram for home-based prices in 2023
axs[2, 0].hist(home_based_2023, bins=10, edgecolor='black')
axs[2, 0].set_title('Home-based Price Distribution in 2023')
axs[2, 0].set_xlabel('Price')
axs[2, 0].set_ylabel('Frequency')

# Histogram for median prices in 2023
axs[2, 1].hist(price_2023, bins=10, edgecolor='black')
axs[2, 1].set_title('Median Price Distribution in 2023')
axs[2, 1].set_xlabel('Price')
axs[2, 1].set_ylabel('Frequency')

# Histogram for household incomes in 2018
axs[3, 0].hist(household_income_2018, bins=10, edgecolor='black')
axs[3, 0].set_title('Household Income Distribution in 2018')
axs[3, 0].set_xlabel('Income')
axs[3, 0].set_ylabel('Frequency')

# Histogram for household incomes in 2023
axs[3, 1].hist(household_income_2023, bins=10, edgecolor='black')
axs[3, 1].set_title('Household Income Distribution in 2023')
axs[3, 1].set_xlabel('Income')
axs[3, 1].set_ylabel('Frequency')

plt.tight_layout()
plt.show()

#Print the histograms to a file
fig.savefig('../output_data/center_based_histogram.png')
```

```python

data = {
    'State': ['Tennessee', 'California', 'Texas', 'New York', 'Florida', 'Illinois'],
    'Median Household Income 2018': [50000, 80000, 60000, 70000, 55000, 65000],
    'Median Household Income 2023': [55000, 85000, 65000, 75000, 60000, 70000]
}
median_df = pd.DataFrame(data)

min_income_2018 = median_df['Median Household Income 2018'].min()
max_income_2018 = median_df['Median Household Income 2018'].max()
min_income_2023 = median_df['Median Household Income 2023'].min()
max_income_2023 = median_df['Median Household Income 2023'].max()

min_state_2018 = median_df.loc[median_df['Median Household Income 2018'].idxmin(), 'State']
max_state_2018 = median_df.loc[median_df['Median Household Income 2018'].idxmax(), 'State']
min_state_2023 = median_df.loc[median_df['Median Household Income 2023'].idxmin(), 'State']
max_state_2023 = median_df.loc[median_df['Median Household Income 2023'].idxmax(), 'State']


years = ['2018', '2023']
max_incomes = [max_income_2018, max_income_2023]
min_incomes = [min_income_2018, min_income_2023]
max_states = [max_state_2018, max_state_2023]
min_states = [min_state_2018, min_state_2023]


plt.figure(figsize=(10, 6))
bar_width = 0.35
indices = np.arange(len(years))


plt.bar(indices, max_incomes, bar_width, label='Max Yearly Income', color='blue')
plt.bar(indices + bar_width, min_incomes, bar_width, label='Min Yearly Income', color='red')


for i in range(len(years)):
    plt.text(indices[i], max_incomes[i] + 1000, max_states[i], ha='center', va='bottom')
    plt.text(indices[i] + bar_width, min_incomes[i] + 1000, min_states[i], ha='center', va='bottom')

plt.xlabel('Year')
plt.ylabel('Household Income')
plt.title('Highest and Lowest Median Household Income for 2018 and 2023')
plt.xticks(indices + bar_width / 2, years)
plt.legend()

plt.show()

#Print the bar plot to a file
plt.savefig('../output_data/min_max_income_bar.png')
```
